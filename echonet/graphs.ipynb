## Grouping cardiac circles

```python
f_grouped = data.groupby('Filename')
```

```python
import math

peak_data = pd.DataFrame()
for group_name, df_group in f_grouped:
    circle = []
    group_peaks = df_group[(df_group['ComputerLarge'] == 1) | (df_group['ComputerSmall'] == 1)]
    if group_peaks.shape[0] == 0:
        group_peaks = df_group[(df_group['Size'] == df_group['Size'].min()) | (df_group['Size'] == df_group['Size'].max())]
    
    counter = 0
    frames = group_peaks['Frame'].to_list()
    circle =[]
    middle_frames = []
    for i in range(1,len(frames)):
        diff = frames[i] - frames[i-1]
        step = math.ceil(0.25 * diff)
        for f in range(frames[i-1],frames[i],step):
            middle_frames.append(math.floor(f))
    final_peaks = df_group[(df_group['ComputerLarge'] == 1) | (df_group['ComputerSmall'] == 1) | (df_group['Frame'].isin(middle_frames))]
    if final_peaks.shape[0] > 0:
        if final_peaks['ComputerSmall'].to_list()[0] == 0:
            counter = 0
        else:
            counter = 0.5
        for j in range(0,final_peaks.shape[0]):
            circle.append(counter)
            counter+=0.125
        # final_peaks = final_peaks.append(pd.DataFrame({'Circle':circle}))
        final_peaks['Circle'] = circle
        if peak_data.shape[0] == 0:
            peak_data = final_peaks
        else:
            peak_data = peak_data.append(final_peaks,ignore_index=True)
    else:
        print(group_name)
```
